```python
#Q6
data=pd.read_csv('employees.csv')
print(data.shape[0])
print(data.shape[1])

print(data.info())
print(data.describe())

print(data.head())
data.tail(3)

print(data['Salary'].mean())
print(data['Bonus'].sum())
print(data['Age'].min())
print(data['Rating'].max())

data.sort_values(by="Salary",ascending=False,inplace=True)
data

l=[]
for i in data['Rating']:
    if(i>=4.5):
        l.append("Excellent")
    elif(i>=4.0 and i<4.5):
        l.append("Good")
    else:
        l.append("Average")
data['Category']=l
data

print(data.isnull())

data.rename(columns={'Employee_ID':'ID'},inplace=True)
data

print(data.loc[data['Years of experience']>5])
print(data.loc[data['Department']=='IT'])

a=[]
for i in data['Salary']:
        a.append(10/100*i)
data['Tax']=a
data

data.to_csv('new_employees.csv')
```
