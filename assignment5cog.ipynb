```python
#Q1
import numpy as np
gfg=np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
a=np.sum(gfg)
print(f"the sum of all elements of the matrix is: {a}")
print("sum of elements row wise is: ")
b=np.sum(gfg[0,0:3])
print(b)
c=np.sum(gfg[1,0:3])
print(c)
d=np.sum(gfg[2,0:3])
print(d)
print("sum of elements column wise is: ")
e=np.sum(gfg[0:3,0])
print(e)
f=np.sum(gfg[0:3,1])
print(f)
g=np.sum(gfg[0:3,2])
print(g)
```

```python
#Q2
import numpy as np
array = np.array([10, 52, 62, 16, 16, 54, 453])
print("the sorted array is: ")
c=array.argsort()
array.sort()
print(array)
print(f"indices are: {c}")
print(f"the four smallest elements are: {array[:4]}")
print(f"the five largest elements are: {array[-5:]}")
d=np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
print("integer elements in this array are: ")
for i in d:
    if i.is_integer():
        print(int(i))
print("floating elements in this array are: ")
for i in d:
    if not i.is_integer():
        print(i)
```

```python
#Q3
name="AS"
sum=0
for i in name:
  sum=sum+ord(i)
print('Sum: ',sum)
sales=np.array([(sum+i*50) for i in range(0,5)])
print('Sales: ',sales)
taxrate=((sum%5)+5)/100
tax=taxrate*sales
print('Tax: ',tax)

#Apply tax
sales=tax+sales
print('Sales with tax: ',sales)
discounted=np.where(sales<(sum+100),sales*0.95,sales*0.9)
print('Sales after discount: ',discounted)

#weeksales
weeklySales=np.tile(discounted,(3,1))
increase=[[1],[1.02],[1.0404]]
print("Weekly Sales:\n",weeklySales*increase)
```

```python
#Q4
import matplotlib.pyplot as plt
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
plt.title('y=x^2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()

y=np.sin(x)
plt.plot(x,y)
plt.title('y=sin(x)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()

y=np.exp(x)
plt.plot(x,y)
plt.title('y=e^x')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()

y=np.log(abs(x)+1)
plt.plot(x,y)
plt.title('y=log(|x|+1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```
