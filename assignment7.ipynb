```python
import numpy as np
import pandas as pd
r_no = 102317243
np.random.seed(r_no)

sales_data = np.random.randint(1000, 5001, size=(12, 4))

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
columns = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]

sales = pd.DataFrame(sales_data, index=months, columns=columns)

print(sales)
```

```python
print(sales.head())
print(sales.describe())

total_sales_per_category = sales.sum()
print()
print(total_sales_per_category)

total_sales_per_month = sales.sum(axis=1)
print()
print(total_sales_per_month)

average_growth = sales.pct_change().mean()
print()
print(average_growth)

sales["Total Sales"] = total_sales_per_month
sales["Growth Rate"] = sales["Total Sales"].pct_change() * 100

if(r_no%2==0):
  sales["Electronics"] *= 0.9
else:
  sales["Electronics"] *= 0.85

print(sales)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=sales[columns], markers=True)
plt.title("Monthly Sales Trends for Each Category")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(columns)
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=sales[columns])
plt.title("Sales Distribution for Each Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
array = np.array([[1, -2, 3], [-4, 5, -6]])

absolute_array = np.abs(array)
print("Absolute Array:\n", absolute_array)

flattened_percentiles = np.percentile(array, [25, 50, 75])
column_percentiles = np.percentile(array, [25, 50, 75], axis=0)
row_percentiles = np.percentile(array, [25, 50, 75], axis=1)
print("Flattened Percentiles:", flattened_percentiles)
print("Column Percentiles:\n", column_percentiles)
print("Row Percentiles:\n", row_percentiles)

flattened_mean = np.mean(array)
flattened_median = np.median(array)
flattened_std = np.std(array)
print("Flattened Mean:", flattened_mean)
print("Flattened Median:", flattened_median)
print("Flattened Standard Deviation:", flattened_std)

column_mean = np.mean(array, axis=0)
column_median = np.median(array, axis=0)
column_std = np.std(array, axis=0)
print("Column Mean:\n", column_mean)
print("Column Median:\n", column_median)
print("Column Standard Deviation:\n", column_std)

row_mean = np.mean(array, axis=1)
row_median = np.median(array, axis=1)
row_std = np.std(array, axis=1)
print("Row Mean:\n", row_mean)
print("Row Median:\n", row_median)
print("Row Standard Deviation:\n", row_std)
```

```python
def swap(lst, index1, index2):
    temp = lst[index1]
    lst[index1] = lst[index2]
    lst[index2] = temp
    return lst

my_list = [1, 2, 3, 4, 5]
index1, index2 = 2, 3
print("Original List:", my_list)
print("Swapped List:", swap(my_list, index1, index2))
def swap(s, elem1, elem2):
    lst = list(s)
    index1, index2 = lst.index(elem1), lst.index(elem2)
    lst[index1], lst[index2] = lst[index2], lst[index1]
    return set(lst)

my_set = {1, 2, 3, 4, 5}
elem1, elem2 = 2, 3
print("Original Set:", my_set)
print("Modified Set:", swap(my_set, elem1, elem2))
```
